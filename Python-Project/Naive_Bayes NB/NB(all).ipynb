```python
import pandas as pa
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
pa.set_option('display.max_rows',None)
```

```python
df = pa.read_excel("fb_comments.xlsx")
```

```python
import nltk

commentaire = df['Comment']
for comment in commentaire:
    tokens = nltk.word_tokenize(comment)
    print(tokens)  # Résultat : ["Je", "suis", "triste"]
```

```python
vectorizer = CountVectorizer(min_df=0, lowercase=False)
```

```python
vec = vectorizer.fit_transform(df['Comment'])
```

```python
vec = np.array(vec.todense())
```

```python
x = vec
y = df['Degout']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

```python
model_NB = MultinomialNB()
```

```python
model_NB.fit(x_train , y_train)
```

```python
y_pred = model_NB.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy : ",accuracy)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

def supprimer_stopwords(commentaire):
    mots = word_tokenize(commentaire)
    mots_sans_stopwords = [mot for mot in mots if mot.lower() not in stopwords.words('arabic')]
    commentaire_sans_stopwords = ' '.join(mots_sans_stopwords)
    return commentaire_sans_stopwords

# Exemple d'utilisation
commentaires = ["هذا تعليق رائع على المقالة", "أشكركم جزيل الشكر على هذه المشاركة", "أنا أوافقكم الرأي"]

commentaires_sans_stopwords = [supprimer_stopwords(commentaire) for commentaire in commentaires]

for commentaire in commentaires_sans_stopwords:
    print(commentaire)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Exemple de commentaires tokenisés
comments = [
    ['ce', 'film', 'est', 'excellent'],
    ['je', 'n\'aime', 'pas', 'cette', 'chanson'],
    ['le', 'livre', 'est', 'captivant'],
    ['je', 'ne', 'recommande', 'pas', 'ce', 'produit'],
    ['لابد', 'تمجيد', 'الطبون'],
['مزلتو', 'القديم', 'ياخي', 'قولتوها', 'عاودتوها'],
['الحمد', 'لله', 'نعمة', 'الجزائر', 'غيرنا', 'يبوسو', 'ليدين', 'الرجلين'],
['بنغلادش', 'نركع', 'نسجد', 'لبشر', 'نقبل', 'الايادي', 'ارجل'],
['جماعة', 'سيدنا', 'ماتعجبهمش', 'هذ', 'الهداير', 'عمي', 'تبون', 'ماتمرڨهاش', 'درك', 'يبدو', 'يعايرو', 'فينا', 'يقولولنا', 'عدنا', 'البنان'],
['بوتفليقة', 'مغربي', 'ولد', 'مدينة', 'وجدة', 'وتخرج', 'جامعاتها']
]

# Vectorisation avec Bag-of-Words
vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform([' '.join(comment) for comment in comments])

# Affichage des mots du vocabulaire
print('Vocabulaire (Bag-of-Words):', vectorizer.vocabulary_)

# Affichage des vecteurs de commentaires
print('Vecteurs de commentaires (Bag-of-Words):')
print(X_bow.toarray())

# Vectorisation avec TF-IDF
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform([' '.join(comment) for comment in comments])

# Affichage des mots du vocabulaire
print('Vocabulaire (TF-IDF):', tfidf_vectorizer.vocabulary_)

# Affichage des vecteurs de commentaires
print('Vecteurs de commentaires (TF-IDF):')
print(X_tfidf.toarray())
```
