```python
import pandas as pa
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
pa.set_option('display.max_rows',None)
```

```python
df = pa.read_excel("fb_comments.xlsx")
```

```python
import nltk

commentaire = df['Comment']
for comment in commentaire:
    tokens = nltk.word_tokenize(comment)
    print(tokens)  # Rรฉsultat : ["Je", "suis", "triste"]
```

```python
vectorizer = CountVectorizer(min_df=0, lowercase=False)
```

```python
vec = vectorizer.fit_transform(df['Comment'])
```

```python
vec = np.array(vec.todense())
```

```python
x = vec
y = df['Degout']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42) 
```

```python
model_NB = MultinomialNB()
```

```python
model_NB.fit(x_train , y_train)
```

```python
y_pred = model_NB.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy : ",accuracy)
```
