```python
import matplotlib.pyplot as plt

# Sample data
x = [407 , 815 , 1223 , 1631 , 2199 , 2299]
y1 = [0.99 , 0.96 , 0.96 , 0.95 , 0.95 , 0.97]
y2 = [0.52 , 0.71 , 0.73 , 0.72 , 0.78 , 0.75]
y3 = [0.67 , 0.64 , 0.65 , 0.63 , 0.67 , 0.65]
y4 = [0.89 , 0.86 , 0.80 , 0.65 , 0.74 , 0.67]
y5 = [0.59 , 0.57 , 0.67 , 0.56 , 0.63 , 0.65]
y6 = [0.87 , 0.84 , 0.83 , 0.83 , 0.90 , 0.87]
y7 = [0.60 , 0.69 , 0.72 , 0.69 , 0.78 , 0.79]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12,6))

# Plot 1
ax.plot(x, y1, label='peur')

# Plot 2
ax.plot(x, y2, label='Tristesse')

# Plot 3
ax.plot(x, y3, label='Colere')

# Plot 4
ax.plot(x, y4, label='Joie')

# Plot 5
ax.plot(x, y5, label='Degout')

# Plot 6
ax.plot(x, y6, label='Surprise')

# Plot 7
ax.plot(x, y7, label='Decu')

# Set labels and title
ax.set_xlabel('Nombre des commentaires')
ax.set_ylabel('précision')
ax.set_title('Graphe de la précision de chaque sentiment par rapport au nombre\ndes commentaires en utilisant l\'algorithme de Classification Decision Tree')

# Create the legend outside the plot
ax.legend(bbox_to_anchor=(1, 1))

# Save the plot to an image png
plt.savefig('common_graph.png', dpi=100)

# Show the plot
plt.show()
```
