```python
import matplotlib.pyplot as plt

# Sample data
x = [407 , 815 , 1223 , 1631 , 2199 , 2299]
y1 = [1.00 , 0.97 , 0.97 , 0.96 , 0.96 , 0.97]
y2 = [0.63 , 0.71 , 0.76 , 0.76 , 0.77 , 0.77]
y3 = [0.50 , 0.65 , 0.63 , 0.63 , 0.65 , 0.70]
y4 = [0.91 , 0.87 , 0.84 , 0.69 , 0.74 , 0.70]
y5 = [0.49 , 0.63 , 0.74 , 0.57 , 0.62 , 0.63]
y6 = [0.84 , 0.85 , 0.87 , 0.87 , 0.91 , 0.89]
y7 = [0.72 , 0.73 , 0.76 , 0.77 , 0.80 , 0.83]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12,6))

# Plot 1
ax.plot(x, y1, label='peur')

# Plot 2
ax.plot(x, y2, label='Tristesse')

# Plot 3
ax.plot(x, y3, label='Colere')

# Plot 4
ax.plot(x, y4, label='Joie')

# Plot 5
ax.plot(x, y5, label='Degout')

# Plot 6
ax.plot(x, y6, label='Surprise')

# Plot 7
ax.plot(x, y7, label='Decu')

# Set labels and title
ax.set_xlabel('Nombre des commentaires')
ax.set_ylabel('précision')
ax.set_title('Graphe de la précision de chaque sentiment par rapport au nombre\ndes commentaires en utilisant l\'algorithme de Classification Random Forest')

# Create the legend outside the plot
ax.legend(bbox_to_anchor=(1, 1))

# Save the plot to an image png
plt.savefig('common_graph.png', dpi=100)

# Show the plot
plt.show()
```
