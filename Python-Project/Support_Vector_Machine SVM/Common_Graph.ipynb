```python
import matplotlib.pyplot as plt 
```

```python
# sample data
x = [407 , 815 , 1223 , 1631 , 2199 , 2299]
y1 = [1.00 , 0.97 , 0.98 , 0.97 , 0.95 , 0.97]
y2 = [0.61 , 0.69 , 0.76 , 0.76 , 0.78 , 0.76]
y3 = [0.51 , 0.67 , 0.68 , 0.66 , 0.66 , 0.68]
y4 = [0.89 , 0.86 , 0.83 , 0.81 , 0.84 , 0.80]
y5 = [0.49 , 0.61 , 0.68 , 0.62 , 0.62 , 0.65]
y6 = [0.83 , 0.83 , 0.83 , 0.85 , 0.89 , 0.87]
y7 = [0.73 , 0.72 , 0.69 , 0.73 , 0.78 , 0.83]

# create a figure and axis
fig , ax = plt.subplots(figsize=(12,6))

# Plot 1 
ax.plot(x , y1 , label = 'peur')

# Plot 2
ax.plot(x , y2 , label = 'Tristesse')

# Plot 3
ax.plot(x , y3 , label = 'Colere')

# Plot 4
ax.plot(x , y4 , label = 'Joie')

# Plot 5
ax.plot(x , y5 , label = 'Degout')

# Plot 6 
ax.plot(x , y6 , label = 'Surprise')

# Plot 7 
ax.plot(x , y7 , label = 'Decu')

# set labels and title 
title = 'Graphe de la précision de chaque sentiment par rapport au nombre \ndes commentaires en utilisant l\'algorithme de Classification SVM'
ax.set_xlabel('Nombre des commentaires')
ax.set_ylabel('précision')
ax.set_title(title)

# Create the legend outside the plot
ax.legend(bbox_to_anchor=(1, 1))

# save the plot to an image png
plt.savefig('common_graph.png' , dpi=100)

# show the plot
plt.show()
```
