```python
# Importation des bibliothèques nécessaires
import pandas as pa
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# afficher toutes les lignes de notre fichier
pa.set_option('display.max_rows',None)
```

```python
# fichier qui contient Données d'entraînement
data_g = pa.read_excel("fb_comments.xlsx")
```

```python
# Conversion des phrases en vecteurs de fonctionnalités numériques
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(data_g['Comment'])
Y = data_g['peur']
```

```python
# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Création du modèle SVM
model = svm.SVC(kernel='linear')
```

```python
# Entraînement du modèle
model.fit(X_train, y_train)
```

```python
# Prédiction sur l'ensemble de test
y_pred = model.predict(X_test)
```

```python
# Affichage des résultats
print(classification_report(y_test, y_pred))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("accuracy : ",accuracy)
```
